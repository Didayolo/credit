#### Credit scoring, predicting the probability that somebody will experience financial distress in the next two years. 

```python
# Import

# Scientific calculation
import numpy as np
from __future__ import division
import random

# Data manipulation
import pandas as pd
import csv

# Data visualization
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

# sklearn, etc.
#from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, roc_auc_score
# Classifiers
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
```

### Data reading

```python
def get_data(f):
    """ Open file f and return dataframe """
    data = pd.read_csv(f)
    data['MonthlyIncome'] = data['MonthlyIncome'].fillna(0) # a mean may be better
    data['NumberOfDependents'] = data['NumberOfDependents'].fillna(0.0)
    data = data.drop(['Unnamed: 0'], axis = 1)
    return data
    
data = get_data('cs-training.csv')
data_test = get_data('cs-test.csv')

y = data['SeriousDlqin2yrs'] # class
X = data.drop(['SeriousDlqin2yrs'], axis = 1)
X_test = data_test.drop(['SeriousDlqin2yrs'], axis = 1)

display(X) #data)
print(len(X))

# stats, distributions ...
```

```python
data.RevolvingUtilizationOfUnsecuredLines.hist(bins=50)
dd = data.loc[data['SeriousDlqin2yrs'] == 0]
sns.distplot(dd['RevolvingUtilizationOfUnsecuredLines'])
plt.title("Distribution de l'attribut DeptRatio chez les non riskys")
plt.show()

sns.jointplot(x="RevolvingUtilizationOfUnsecuredLines", y="SeriousDlqin2yrs", data=data)
print("Distribution des classes en fonction de l'attribut RevolvingUtilizationOfUnsecuredLines")
plt.show()
```

```python
# Sampling, cross-validation
samp = 100000
x_train, x_validation, y_train, y_validation = train_test_split(X[:samp], y[:samp], test_size=0.2, random_state=0)
```

### Classification, validation...

```python
#SC = svm.SVC(kernel='poly', degree=4)
#SC.fit(X[:3000], y[:3000])
#SC.score(X[3000:3500], y[3000:3500])

# Validation
DT = DecisionTreeClassifier()

DT.fit(x_train.values, y_train.values)
print(DT.score(x_validation.values, y_validation.values))

# Classification
clf = QuadraticDiscriminantAnalysis()
clf.fit(X, y)
pred = clf.predict_proba(X_test)

#false_positive_rate, true_positive_rate, thresholds = roc_curve(y, clf.predict_proba(X)[:,1])
#print roc_auc_score(y, clf.predict(X))
```

```python
CL = DT
importances = list(zip(CL.feature_importances_, x_train.columns))
importances.sort(reverse=True)
pd.DataFrame(importances, index=[x for (_,x) in importances]).plot(kind = 'bar', color='red')
plt.show()
```

### Output for Kaggle

```python
with open('submission.csv', 'w') as csvfile:
    fieldnames = ['Id', 'Probability']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    for i in range(len(pred)):
        writer.writerow({'Id': i+1, 'Probability': pred[i][1]}) # proba membership 1
```
